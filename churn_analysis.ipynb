```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
print('Data loaded!', df.shape)
```

```
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
df['Churn_Flag'] = df['Churn'].map({'Yes': 1, 'No': 0})
print('Churn Rate:', round(df['Churn_Flag'].mean() * 100, 2), '%')
```

```
churn_counts = df['Churn'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(churn_counts, labels=['Stayed','Churned'], autopct='%1.1f%%', colors=['#1E8449','#E74C3C'], startangle=90, explode=(0, 0.1))
plt.title('Customer Churn Rate', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('churn_rate.png', dpi=150)
plt.show()
```

```
contract_churn = df.groupby('Contract')['Churn_Flag'].mean() * 100
print(contract_churn)
plt.figure(figsize=(8,5))
contract_churn.plot(kind='bar', color=['#E74C3C','#F39C12','#1E8449'])
plt.title('Churn Rate by Contract Type', fontsize=14, fontweight='bold')
plt.ylabel('Churn Rate (%)')
plt.xlabel('Contract Type')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('contract_churn.png', dpi=150)
plt.show()
```

```
plt.figure(figsize=(9,5))
df[df['Churn']=='Yes']['MonthlyCharges'].hist(bins=30, alpha=0.7, color='#E74C3C', label='Churned')
df[df['Churn']=='No']['MonthlyCharges'].hist(bins=30, alpha=0.7, color='#1E8449', label='Stayed')
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Count')
plt.title('Monthly Charges: Churned vs Stayed', fontsize=13, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.savefig('charges_distribution.png', dpi=150)
plt.show()
```

```
plt.figure(figsize=(9,5))
df[df['Churn']=='Yes']['tenure'].hist(bins=30, alpha=0.7, color='#E74C3C', label='Churned')
df[df['Churn']=='No']['tenure'].hist(bins=30, alpha=0.7, color='#1E8449', label='Stayed')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count')
plt.title('Tenure: Churned vs Stayed', fontsize=13, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.savefig('tenure_churn.png', dpi=150)
plt.show()
```

```
import sqlite3
conn = sqlite3.connect('churn.db')
df.to_sql('customers', conn, if_exists='replace', index=False)
query1 = '''
SELECT Contract, COUNT(*) AS total_customers, SUM(Churn_Flag) AS churned, ROUND(AVG(Churn_Flag)*100, 1) AS churn_rate_pct
FROM customers GROUP BY Contract ORDER BY churn_rate_pct DESC
'''
print(pd.read_sql(query1, conn))
conn.close()
```
